```python
import pandas as pd
import numpy as np
from pprint import pprint
import matplotlib.pyplot as plt
import os
from scipy.interpolate import make_interp_spline
```

```python
log_df = pd.read_csv(r"C:\Users\parzi\OneDrive - Tribhuvan University\Desktop\Minor Project\Monument Detection with CNN\Monument Object Detection\Evaluations\log\test_map_score.csv")
```

```python
font_title = {'size': 14, 'weight':'bold'}
font_label = {'size': 12, 'weight':'bold'}
plt.figure(figsize = (8, 5), dpi = 90)
plt.style.use('seaborn-whitegrid')

plt.plot(log_df['map score'])
plt.xlabel('Number of epochs', fontdict = font_label)
plt.ylabel('mAP Score (%)', fontdict = font_label)
plt.title('mAP Score Curve for Test Dataset', fontdict = font_title)

ymin = list(log_df['map score'])[0]
ymax = list(log_df['map score'])[-1]

stepsize = (ymax - ymin) / 15.0
plt.yticks(np.arange(ymin, ymax + stepsize, stepsize))
plt.xticks(np.arange(0, 60, 5))

saveloc = r"C:\Users\parzi\OneDrive - Tribhuvan University\Desktop\Minor Project\Results and Analysis\Offline Model\mapPlot.png"
plt.savefig(saveloc, dpi = 120, bbox_inches = 'tight', pad_inches = 0.15)
plt.show()
```
